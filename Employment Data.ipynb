```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Create a reference the CSV file desired
employment_csv = "Employment Data/Employment Summary Data.csv"

# Read the CSV into a Pandas DataFrame
employment_df = pd.read_csv(employment_csv)

# Print the first five rows of data to the screen
employment_df.head()
```

```python
employment_df.dtypes
```

```python
#time series line plots for unemployment rate
xu=employment_df["Date"]
yu=employment_df["Unemployment Rate"]
plt.plot(xu,yu)
plt.plot(x_compat=True)
```

```python
#time series line plots for labor force participation
xp=employment_df["Date"]
yp=employment_df["Labor Force Participation"]
plt.plot(xp,yp)
```

```python
#time series line plots for initial jobless claims
xc=employment_df["Date"]
yc=employment_df["Initial Jobless Claims"]
plt.plot(xc,yc)
```

```python
#time series line plots for job openings
xo=employment_df["Date"]
yo=employment_df["Job Openings"]
plt.plot(xo,yo)
```
