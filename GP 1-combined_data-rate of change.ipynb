```python
# Import Dependencies
%pylab inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.dates as mdates
```

```python
# Use Pandas data frame to read csv data and read the first data
enrollmentRate_df1 = pd.read_csv("enrollmentRateByYear.csv")
unemploymentRate_df1 = pd.read_csv("unemploymentRateByYear.csv")
completionRate_df1 = pd.read_csv("completionRateByYear.csv")
```

## Enrollment data

```python
enrollmentRate_df1.head()
```

```python
# Rename Columns in Enrollment data set
enrollmentRate_df2 = enrollmentRate_df1.rename(columns = {"Enrollment Annual Rate of Change":"Enrollment RateOfChange"})
enrollmentRate_df2.head()
```

```python
# Select Desired Columns in FINAL Enrollment data set
enrollmentRate_df3 = enrollmentRate_df2[['year','Enrollment RateOfChange']]
enrollmentRate_df3.head()
```

## Employment data

```python
unemploymentRate_df1.head()
```

```python
# Select Desired Columns in Employment data set
unemploymentRate_df2 = unemploymentRate_df1[['Year','Unemployment Annual Rate of Change']]
unemploymentRate_df2.head()
```

```python
# Rename Columns in Employment data set
unemploymentRate_df3 = unemploymentRate_df2.rename(columns = {"Year":"year", "Unemployment Annual Rate of Change":"Unemployment RateOfChange"})
unemploymentRate_df3.head()
```

## Completion data

```python
completionRate_df1.head()
```

```python
# Select Desired Columns in Completion data set
completionRate_df2 = completionRate_df1[["year","Completion Annual Rate of Change"]]
completionRate_df2.head()
```

```python
#COMPLETION: Rename
completionRate_df3 = completionRate_df2.rename(columns = {"Completion Annual Rate of Change":"Completion RateOfChange"})
completionRate_df3.head()
```

## Combine 3 datasets for analysis

```python
combinedRates_df1 = pd.merge(pd.merge(enrollmentRate_df3, unemploymentRate_df3, on="year"), completionRate_df3, on="year")
combinedRates_df1.head()
```

```python
#Table was used in Technical Writting Summary
combinedRates_df2 = combinedRates_df1.set_index("year")
combinedRates_df2
```

```python
multi_plot = combinedRates_df2.plot(kind="bar", figsize=(20,5), color=['coral', 'aqua', 'blue'])
multi_plot.set_xticklabels(combinedRates_df2.index, rotation=45)
plt.ylim(-17, 65)
plt.title("Combined Rate of Change in Bar chart (2005-2017)")
plt.savefig("Combined Rate of Change in Bar chart (2005-2017)")
plt.ylabel("Annual Rate of change")
plt.show()
plt.tight_layout()
```

```python
#ZOOMED IN *******************************************
multi_plot = combinedRates_df2.plot(kind="bar", figsize=(20,5), color=['coral', 'aqua', 'blue'])
multi_plot.set_xticklabels(combinedRates_df2.index, rotation=45)
plt.ylim(-2, 7)
plt.title("Combined Rate of Change in Bar chart (2005-2017)")
plt.savefig("Zoomed in Combined Rate of Change in Bar chart (2005-2017)")
plt.ylabel("Annual Rate of change")
plt.show()
plt.tight_layout()
```
